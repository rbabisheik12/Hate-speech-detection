```python
import gensim
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Define stop words (modify as needed)
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove punctuation
    text = ''.join([char for char in text if char.isalnum() or char.isspace()])

    # Remove stop words
    tokens = word_tokenize(text)
    tokens = [token for token in tokens if token not in stop_words]


    return tokens
```

```python
import os

def combine_text_files(folder_path):

    no_of_files = 0

    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, "r") as file:
                text = file.read()

            processed_tokens = preprocess_text(text)
            my_string = ' '.join(processed_tokens)

            new_file_name = "/content/preprocessed txt file/" + filename

            with open(new_file_name, "w") as f:
              f.write(my_string)

            no_of_files += 1

        print(no_of_files)
```

```python
import os
import numpy as np

def convert_to_array(folder_path):

    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, "r") as file:
              print("with opened")
              text = file.read()
              tokens = text.split()  # Tokenize text (simplified)
              word_vectors = []
              for token in tokens:
                try:
                    vector = word2vec_model[token]
                    word_vectors.append(vector)
                except KeyError:
                    # Handle out-of-vocabulary words
                    pass

              word_vectors = np.array(word_vectors)
              # Save word vectors as numpy array
              np.save(os.path.join("/content/np text arrays new", f"{filename[:-4]}.npy"), word_vectors)
```

```python
import os
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
from sklearn.model_selection import train_test_split

class NumpyDataset(Dataset):
    def __init__(self, folder_path, file_names):
        self.folder_path = folder_path
        self.file_names = file_names

    def __len__(self):
        return len(self.file_names)

    def __getitem__(self, idx):
        file_name = self.file_names[idx]
        file_path = os.path.join(self.folder_path, file_name)

        # Load numpy array from file
        features = np.load(file_path)

        # Check if the array is empty
        if features.shape[0] == 0:
            # Replace with a small random tensor
            features = torch.randn(1, 300)
        else:
            features = torch.from_numpy(features)

        # Extract label from file name
        label = int(file_name[0] ) # Remove last four characters (.npy extension)

        return features, label

# Specify folder containing .npy files
folder_path = "/content/np text arrays new"

# Get list of all .npy files in the folder
file_names = [file for file in os.listdir(folder_path) if file.endswith(".npy")]

# Split dataset into training and testing sets
train_files, test_files = train_test_split(file_names, test_size=0.2, random_state=42)

# Create dataset instances for training and testing
train_dataset = NumpyDataset(folder_path, train_files)
test_dataset = NumpyDataset(folder_path, test_files)
batch_size = 32

def collate_fn(batch):
    features, labels = zip(*batch)
    features = pad_sequence(features, batch_first=True, padding_value=0)
    return features, labels

train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, collate_fn=collate_fn)
```

```python
import torch.nn as nn

class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(RNNModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)  # Output size is 1 for binary classification
        self.sigmoid = nn.Sigmoid()  # Sigmoid activation for binary classification

    def forward(self, x):
        # Set initial hidden and cell states
        device = x.device
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size, device=device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size, device=device)


        # Forward propagate LSTM
        out, _ = self.lstm(x, (h0, c0))  # out: tensor of shape (batch_size, seq_len, hidden_size)

        # Decode the hidden state of the last time step
        out = out[:, -1, :]
        out = self.fc(out)
        out = self.sigmoid(out)  # Apply sigmoid activation for binary output
        return out.squeeze()  # Remove the last dimension (size 1)
```
